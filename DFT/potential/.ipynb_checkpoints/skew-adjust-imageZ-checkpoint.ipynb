```python
def scatterweight (target,ICISS):

    df = pd.read_table(target)
    X = list(df['rx'])
    Y = list(df['ry'])
    bx,by = max(X)-min(X),max(Y)-min(Y)
    print([bx,by])

    atom = pd.read_table(ICISS,delim_whitespace=True,names = ('x','y','z','Z','m','f'))

    #find the top unblocked atoms
    toplist = []    
    for index, row in atom.iterrows():
        k=0
        for a in toplist:
            rowa = atom.loc[a]
            if np.sqrt((rowa['x']-row['x'])**2+(rowa['y']-row['y'])**2)<0.4:
                if rowa['z']>row['z']:
                    k=1
                else:
                    toplist.remove(a)
        if k==0:
            toplist.append(index)
    print(toplist)

    counts = [0]*atom.shape[0]
    for x,y in zip(X,Y):
        dmin = 500
        index_min = -1
        for a in toplist:
            lx = abs(atom.loc[a]['x']-x)
            ly = abs(atom.loc[a]['y']-y)
            l = min(lx,bx-lx)**2+min(ly,by-ly)**2
            if  l < dmin:
                dmin = l
                index_min = a
        counts[index_min] += 1

    atom['count']=counts



    atom['sum'] = atom.groupby('Z')['count'].transform('sum')
    atom['perc'] = (atom['count']/atom['sum']).round(3)
    atom.head(10)


    topIn = atom.loc[toplist].query('Z ==49')
    topAs = atom.loc[toplist].query('Z ==33')
    In_X = list(topIn['x'])
    In_Y = list(topIn['y'])
    As_X = list(topAs['x'])
    As_Y = list(topAs['y'])
    plt.figure(figsize=(bx/2,by/2))
    plt.scatter(X,Y,marker='.')
    plt.scatter(In_X,In_Y,marker='*',s=100)
    plt.scatter(As_X,As_Y,marker='o',s=150)
    
    return topIn['perc'],topAs['perc']
    
```

```python
def band(listZ,Zcutoff):
    g1= np.clip(Cmax-C*np.exp(-(Zcutoff-listZ)/w),Cmax-C,Cmax)
    return g1
```

```python
class Atom:
    def __init__(self,atomindex,termination,azi):
        self.atomIndex = atomindex
        self.azi = azi
        self.theta = theta
        self.offset = Parlist[termination]['offset']
        self.file =Parlist[termination]['File']
        self.symbol = self.file.symbols[self.atomIndex]
        self.position = self.file.atomxyz[self.atomIndex]
        self.data = self.file.data
        self.x = self.position[0]
        self.y = self.position[1]
        self.z = self.position[2]
        self.mass = M[self.symbol]
        self.abovepotential = self.potential(begin=self.z,end=cellZ,split=psplit)
        self.Z0 = self.Z0()#self.Zcutoff()-2#self.file.maxZ-#
        self.cx = math.sin(self.theta/180*np.pi)*math.cos(self.azi/180*np.pi)
        self.cy = math.sin(self.theta/180*np.pi)*math.sin(self.azi/180*np.pi)
        self.cz = math.cos(self.theta/180*np.pi)
        self.fermi =Parlist[termination]['fermi']
        self.Z0potential = self.potential(begin=self.Z0,end=self.Z0+4)
        self.Zcutoff = self.Zcutoff()
        self.NF = self.NF()
        
            
    
    def Z0(self):
        Z0 = self.z
#         for ip,p in enumerate(self.abovepotential):
#             if min(self.abovepotential[ip:])<-40:
#                 Z0+=(cellZ-self.z)/psplit
#             else:
#                 return Z0
        return Z0+3

    def Zcutoff(self):    
        Zcuttoff=self.z
        for [x,y,z] in self.file.atomxyz:
            if min(abs(x-self.x),cellX-abs(x-self.x))**2+min(abs(y-self.y),cellY-abs(y-self.y))**2<=25:
                Zcuttoff= max(Zcuttoff,z)
        return Zcuttoff+4
        
    #calculate freezing distancce of 3000eV Na scattered from InAs at 125
    #the projectile comes out in surface normal direction 
    
    def read_data(self,iX,iY,iZ):
        #convert XYZ to within boundary first
        while  iX<0: 
            iX += boundary[0] 
        while  iX>=boundary[0]: 
            iX -= boundary[0]
        while  iY<0: 
            iY += boundary[1] 
        while  iY>=boundary[1]: 
            iY -= boundary[1]
        iZ = min(iZ,boundary[2]-1)
        return self.data[iX][iY][iZ]

    def xyzpotential(self,X,Y,Z):
        #p(X,Y,Z) = p(int(X,Y,Z))+dp/d(x,y,z)*p(rx,ry,rz)
        iX,iY,iZ = int(X/dx),int(Y/dy),int(Z/dz)
        decimal = Z%dz
        gradient = (self.read_data(iX,iY,iZ+1)-self.read_data(iX,iY,iZ))/dz
        p = self.read_data(iX,iY,iZ)+gradient*decimal

        return round(p-self.offset,3) 
    
    def potential(self,begin,end,split=20,grid=9):
        XYpairs=[]
        
        for delta_x in np.linspace(-r,r,grid):
            for delta_y in np.linspace(-r, r, grid):
                if np.abs(delta_x+delta_y*1.j)<=r:
                    XYpairs+=[(self.x+delta_x,self.y+delta_y)]
        
        self.avg_p_z=[]
        for ddz in np.linspace(begin,end,split): 
            p_tot = []
            for PX,PY in XYpairs:
                p_tot.append(self.xyzpotential(PX,PY,ddz))
            self.avg_p_z+=[round(np.mean(p_tot),2)]       
        return self.avg_p_z 
    
    
    #use RCT model to determine NF scattered from each site
    def NF(self):
        
        factor = 1 if self.symbol=='In' else 1
        Z = Zrange
        
        self.g1=band(Zrange+self.Z0,self.Zcutoff)
        #calculate s level after shifting 
        dE = 27.2/np.sqrt((27.2*27.2)/(vmax*vmax) + 16*((Z)/0.53)**2)
        self.Es = -5.14+dE+self.Z0potential
        #Es = self.Es
        #calculating broadening of the s level
        broad = 27.2*2.23/np.sqrt(np.sqrt(np.exp(4*0.86*Z/0.53)+(2.23/0.04)**4))*factor
        #print(broad)
        
        #guassion integral
        sigma = broad/1.38629
        NF = (0.5+0.5*np.erf((self.fermi-self.Es)/(1.414*sigma)))-(0.5+0.5*np.erf((self.g1-self.Es)/(1.414*sigma))-0.5-0.5*np.erf((self.g1-G-self.Es)/(1.414*sigma)))
        return np.round(NF*100,1) #NF in %

    def NF_plot(self):
        if self.symbol == 'As':
            plt.plot(Zrange,self.NF,'-',label = 'As'+str(self.atomIndex))
        else:
            plt.plot(Zrange,self.NF,'--',label = 'In'+str(self.atomIndex))
        #plt.legend(loc='best')
        
```

```python
def FreezingDistance(m2,E0,sc):
    m1 = 22.99
    E = E0*((m1*np.cos(sc/180*np.pi)+np.sqrt(m2**2-(m1*np.sin(sc/180*np.pi))**2))/(m1+m2))**2
    #NA=6.02*(10^23)   #Avogadora 
    #m=22.989769/NA/1000 #Na mass in kg
    #E=E*1.60217662*(10^-19) #Energy in Joul
    #v=math.sqrt(2*E/m)=math.sqrt(2*E*1.602/22.99*6.02*10000000)
    v = math.sqrt(8389769*E)*math.cos(theta/180*np.pi)
    Zf = round(1/0.86*math.log(2*2.23/(0.86*v/2187691))*0.53,2)
    return Zf
```

```python
def NF_avg_plot(termination,azi,Symbol,linestyle,linewidth):
    
    weightlist = Parlist[termination][azi][Symbol]['ICISS']
    NFs = []
    for i in weightlist.index:
        NFs += [Atom(i,termination,azi).NF*weightlist.loc[i]]
    NF_avg = np.array(NFs).sum(axis=0)   
    plt.plot(Zrange,NF_avg,linestyle=linestyle,linewidth=linewidth,label=Symbol+'_avg')
    plt.legend(loc='best')
```

```python
def exp_plot(termination,azi,Symbol,color,linestyle,alpha=0.1):
    expnf = Parlist[termination][azi][Symbol]['expNF']
    experr = Parlist[termination][azi][Symbol]['err']
    plt.axhline(y=expnf,color=color,linestyle=linestyle)
    plt.axhspan(expnf-experr,expnf+experr,color=color,linestyle=linestyle,alpha=alpha)
    plt.axvline(x=Zf[Symbol],color=color,linestyle=linestyle)
```
